```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_aapl = pd.read_json("Stock-aapl.json").chart
df_googl = pd.read_json("Stock-googl.json").chart

def getPrice (df):
    n = len(df)
    price = np.zeros(n)
    for i in range (n):
        price[i] = df.iloc[i]['close']
    return price

price_aapl = getPrice (df_aapl)
price_googl = getPrice (df_googl)

def plotPrice (price , name):
    plt.figure(figsize=(16,8))
    plt.title(name)
    plt.plot (range(len(price)) , price)
    plt.show()
    
    
```

```python
plotPrice (price_aapl , "AAPL")
plotPrice (price_googl , "G")
```

```python
def slidingCorrl (price1 , price2 , k=100):
    n = len(price1)
    correl = np.zeros(n-k+1)
    for i in range (n-k+1):
        correl[i] = np.corrcoef (price1 [i:i+k-1] , price2 [i:i+k-1])[0][1]
    return correl
```

```python
correl = slidingCorrl (price_aapl , price_googl)
```

```python
plotPrice (correl , "Correl AAPL-GOOGL")
```
